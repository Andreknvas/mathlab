```python
%matplotlib widget

import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

dt = 0.01
num_steps = 15000

def lorenz(x, y, z, s=10, r=28, b=2.667):
    return (s*(y - x), r*x - y - x*z, x*y - b*z)

xs = np.empty(num_steps)
ys = np.empty(num_steps)
zs = np.empty(num_steps)
xs[0], ys[0], zs[0] = 1.0, 1.0, 1.0

for i in range(num_steps - 1):
    dx, dy, dz = lorenz(xs[i], ys[i], zs[i])
    xs[i+1] = xs[i] + dx * dt
    ys[i+1] = ys[i] + dy * dt
    zs[i+1] = zs[i] + dz * dt


speed = np.sqrt(np.diff(xs)**2 + np.diff(ys)**2 + np.diff(zs)**2)
speed = (speed - speed.min()) / (speed.max() - speed.min())

fig = plt.figure(figsize=(11, 8))
ax = fig.add_subplot(111, projection='3d')

for i in range(0, num_steps - 1, 3):
    ax.plot(xs[i:i+4], ys[i:i+4], zs[i:i+4],
            color=plt.cm.inferno(speed[i]), lw=0.4, alpha=0.7)

ax.set_axis_off()
ax.set_facecolor('black')
fig.patch.set_facecolor('black')
ax.set_title('Atrator de Lorenz', color='white', fontsize=14, pad=20)

plt.tight_layout()
plt.show()
```

```python
%matplotlib widget

import numpy as np
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

u = np.linspace(0, 2 * np.pi, 200)
v = np.linspace(0, np.pi, 200)
U, V = np.meshgrid(u, v)

X = (2 + np.cos(V / 2) * np.sin(U) - np.sin(V / 2) * np.sin(2 * U)) * np.cos(V)
Y = (2 + np.cos(V / 2) * np.sin(U) - np.sin(V / 2) * np.sin(2 * U)) * np.sin(V)
Z = np.sin(V / 2) * np.sin(U) + np.cos(V / 2) * np.sin(2 * U)

ax.plot_surface(X, Y, Z, cmap='plasma', alpha=0.9, linewidth=0)
ax.set_axis_off()
ax.set_title('Garrafa de Klein', fontsize=14, pad=20)

plt.tight_layout()
plt.show()
```
